```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import pandas as pd

data = {
    'Days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Study_Hours': [8, 6, 7, 5, 7],
    'Marks': [90, 85, 92, 78, 88],
    'Attendance': [95, 90, 92, 88, 95],
    'Sleep_Hours': [7, 6, 8,7,8],
    'Assignments_Completed': [5, 4, 6, 3, 5]
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df['Days'], df['Study_Hours'], marker='o')
plt.title('Study Hours Trend')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Days'], df['Study_Hours'], marker='o', color='blue', label='Study Hours')
plt.plot(df['Days'], df['Marks'], marker='o', color='green', label='Marks')
plt.title('Study Hours and Marks Over Time')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
```

```python
plt.bar(df['Days'], df['Marks'],color= 'turquoise')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
bars = plt.bar(df['Days'], df['Marks'], color='red')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y),ha='center', va='bottom')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y') # adds horizontal grid for readability
plt.show()
```

```python
#4. Scatter Plot (Relationship)
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Study_Hours': [8, 6, 7, 5, 7],
    'Marks': [90, 85, 92, 78, 88],
    'Attendance': [95, 90, 92, 88, 95],
    'Sleep_Hours': [7, 6, 8,7,8],
    'Assignments_Completed': [5, 4, 6, 3, 5]
}
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'],df['Marks'],color='green')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Condtitonal Scatter plot
df['Result'] = ['Fail','Pass','Pass','Fail','Pass']

colors = {'Fail':'Red','Pass':'Black'}
color_map = df['Result'].map(colors)
plt.scatter(df['Study_Hours'],df['Marks'],c=color_map)
plt.title('Study Hours vs Marks (result -wise)')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

SAME GRAPHS USING SEABORN (BETTER VISUALISATION)

```python
import seaborn as sns
#Line Plot
sns.lineplot(data=df,x='Days',y='Study_Hours')
plt.title('Study Hours Trend')
plt.show()
```

```python
#Histrogram
sns.histplot(data=df,x='Marks',bins=3,color='purple')
plt.title('Marks Distribution')
plt.show()
```
